```python
# Define the file paths
file_path_tweets_eiagov = 'Original Data/X(tweets) data/X(tweets)_EIAgov.csv'
file_path_tweets_us_oil = 'Original Data/X(tweets) data/X(tweets)_U.S.Oil.csv'

# Read the CSV files
df_tweets_eiagov = pd.read_csv(file_path_tweets_eiagov)
df_tweets_us_oil = pd.read_csv(file_path_tweets_us_oil)
```

### 1. Dataset Overview:

**Dataset Description:**
The datasets you are working with are collections of news articles and tweets, each containing sentiment analysis data. These datasets are likely used to analyze how public sentiment reflected in news outlets and social media impacts or correlates with financial markets, specifically crude oil prices. 

### 2. Data Sources:
- **Social Media Dataset (Final_X_articles.csv):** This dataset contains tweets, most likely focused on crude oil or energy-related topics. The tweets have been processed to include sentiment scores and possibly other metadata.
- **How data set was obtained can be found in the "New York Times Web Scraping Code" folder.
- **News Dataset (Final_NYT_articles.csv):** This dataset contains articles from the New York Times. Each row represents a news article with various features related to the content and sentiment of the article.
- **How data set was obtained can be found in the "New York Times Web Scraping Code" folder.


### 3. Key Features (Columns):

#### **(Final_X_articles.csv) is a joined and merged datasets with the search query of ("U.S. Oil" and "EIA Oil"):**
- **Date:** The posting date of each tweet.
- **Tweet Text:** The content of the tweet.
- **User Information:** Metadata like the username, follower count, and location of the person tweeting.
- **Engagement Metrics:** Columns like tweet views, retweets, favorites, or replies could represent the tweet’s engagement level.

#### **NYT Articles Dataset (Final_NYT_articles.csv):**
- **Date:** The publication date of each article.
- **Headline/Abstract:** Text data representing the headline and abstract of the article.





### 4. Feature Representation:

- **Text Features:** Headline, abstract, and tweet text contain the raw text data.
- **Date:** Helps in the time-series analysis to compare the sentiment data against financial indicators like crude oil prices.
- **User Information (in tweets):** Provides insights into the source of information and possibly influencer analysis.

### 5. Metadata:
- **Publication or Posting Date:** Important for time-series analysis to track the progression of sentiment over time.
- **User or Article Metadata:** In the case of articles, metadata like section name, publication, and author provide additional context, while for tweets, information about the user and engagement gives insights into the reach and potential impact of the content.

These datasets will allow for sentiment analysis and correlation studies with external data, such as crude oil prices.

```python
df_tweets_eiagov['tweet.created_at'] = pd.to_datetime(df_tweets_eiagov['tweet.created_at'])
df_tweets_us_oil['tweet.created_at'] = pd.to_datetime(df_tweets_us_oil['tweet.created_at'])

# Merge the DataFrames
merged_df = pd.concat([df_tweets_eiagov, df_tweets_us_oil])

# Sort the merged DataFrame by tweet.created_at
merged_df = merged_df.sort_values(by='tweet.created_at').reset_index(drop=True)

# update the tweet_count column to represent the order of tweets in the merged DataFrame
merged_df['tweet_count'] = range(1, len(merged_df) + 1)

# Display the last few rows of the merged DataFrame
merged_df.tail()
```

```python
# Count the number of tweets that are marked as possibly sensitive
num_sensitive_tweets = merged_df[merged_df['tweet.possibly_sensitive'] == True].shape[0]

num_sensitive_tweets

```

## Removing Possibly Sensitive Column as it has no value

```python
# Remove the 'tweet.possibly_sensitive' column
merged_df.drop(columns=['tweet.possibly_sensitive'], inplace=True)

# Display the first few rows of the updated DataFrame
merged_df.head()
```

```python
# Count the number of tweets that have withheld countries
num_withheld_in_countries = merged_df[merged_df['tweet.user.withheld_in_countries'].apply(lambda x: len(eval(x)) > 0)].shape[0]

num_withheld_in_countries
```

```python

# Display the tweet that has been withheld in countries
withheld_tweet = merged_df[merged_df['tweet.user.withheld_in_countries'].apply(lambda x: len(eval(x)) > 0)]

withheld_tweet
```

### One tweet with the 'RU' (Russia) country code has been withheld. Therefore the witheheld_in_countries column is not nessasary and can be removed as well. 

```python

merged_df.drop(columns=['tweet.user.withheld_in_countries'], inplace=True)


merged_df.head()
```


### Key Insights:

   
1. **tweet.user.followers_count**: 
   - Huge variation in follower counts. The standard deviation is very large (994,035), and the mean is significantly skewed by high-value accounts (e.g., the max is over 20 million).

2. **tweet.view_count**: 
   - Average views per tweet is 10,645, but there's a lot of variation (standard deviation: 70,415).
   - 25% of tweets received fewer than 251 views, and 75% received fewer than 4,568 views. The max view count is over 1.5 million.

3. **tweet.favorite_count**: 
   - Most tweets have very few likes. 50% of tweets have 7 or fewer likes, and 25% have only 1 like. However, there are a few tweets with very high favorite counts, leading to a high standard deviation.

4. **tweet.retweet_count**: 
   - Similar to favorite counts, most tweets have very few retweets, with a median of only 2 retweets. However, a few tweets have thousands of retweets (max 9519).

5. **tweet.reply_count**: 
   - Most tweets have very few replies, with 50% of tweets having zero replies.

This overview shows that there are a few tweets in the dataset with extremely high engagement (views, likes, retweets), while the majority of tweets have low to moderate engagement.

# Let's Visualize distributions: Use histograms, boxplots, or bar charts to visualize numerical and categorical distributions.

```python
Final_X_articles = merged_df


import matplotlib.pyplot as plt

# Set up the figure size
plt.figure(figsize=(16, 10))

# Histogram for numerical columns
numerical_cols = ['tweet.user.followers_count', 'tweet.view_count', 'tweet.favorite_count', 'tweet.retweet_count', 'tweet.reply_count']
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i+1)
    plt.hist(Final_X_articles[col].dropna(), bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()


```

```python
# Bar charts for categorical columns
plt.figure(figsize=(16, 10))

# Bar chart for tweet.user.name (Top 10 most frequent users)
top_users = Final_X_articles['tweet.user.name'].value_counts().nlargest(10)
plt.subplot(2, 1, 1)
top_users.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Top 10 Users by Tweet Count')
plt.xlabel('User Name')
plt.ylabel('Tweet Count')

# Bar chart for tweet.user.location (Top 10 most frequent locations)
top_locations = Final_X_articles['tweet.user.location'].value_counts().nlargest(10)
plt.subplot(2, 1, 2)
top_locations.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 10 Locations by Tweet Count')
plt.xlabel('Location')
plt.ylabel('Tweet Count')

plt.tight_layout()
plt.show()
```

## Here are the bar charts displaying the top 10 users and locations based on tweet counts. These provide a quick view of the most active users and locations in the dataset

```python
# Convert pub_date to datetime for visualization
nyt_data['pub_date'] = pd.to_datetime(nyt_data['pub_date'], errors='coerce')

# Set up the figure size
plt.figure(figsize=(16, 12))

# Bar chart for 'section_name' (Top sections by article count)
plt.subplot(3, 1, 1)
section_counts = nyt_data['section_name'].value_counts()
section_counts.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Distribution of Articles by Section')
plt.xlabel('Section Name')
plt.ylabel('Article Count')

# Bar chart for 'byline' (Top 10 most frequent authors)
plt.subplot(3, 1, 2)
top_bylines = nyt_data['byline'].value_counts().nlargest(10)
top_bylines.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 10 Authors by Article Count')
plt.xlabel('Byline')
plt.ylabel('Article Count')

# Histogram for 'pub_date' (Articles per day)
plt.subplot(3, 1, 3)
nyt_data['pub_date'].dt.date.value_counts().sort_index().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Articles by Publication Date')
plt.xlabel('Publication Date')
plt.ylabel('Article Count')

plt.tight_layout()
plt.show()
```
