```python
import pandas as pd

# Read the CSV files
X_articles = pd.read_csv('Final Cleaned Data with (Sentiment & Classification)/X_articles_results.csv')
X_articles['source'] = 'X'
X_articles.tail()
```

## Creating a column just with the date, without the time so we can sort it later on.

```python
# Convert 'pub_date' to datetime and create 'date' column
NYT_articles['pub_date'] = pd.to_datetime(NYT_articles['pub_date'])
NYT_articles['date'] = NYT_articles['pub_date'].dt.date
cols = list(NYT_articles.columns)
pub_date_index = cols.index('pub_date')
cols.insert(pub_date_index, cols.pop(cols.index('date')))
NYT_articles = NYT_articles[cols]


# Convert 'pub_date' to datetime and sort by 'pub_date'
NYT_articles['pub_date'] = pd.to_datetime(NYT_articles['pub_date'])
NYT_articles = NYT_articles.sort_values(by='pub_date').reset_index(drop=True)
NYT_articles.head(1)

```

```python
# Convert 'tweet.created_at' to datetime and create 'publish date' column
X_articles['tweet.created_at'] = pd.to_datetime(X_articles['tweet.created_at'])
X_articles['date'] = X_articles['tweet.created_at'].dt.date

# Rearrange the columns so that 'publish date' comes before 'tweet.created_at'
cols = list(X_articles.columns)
created_at_index = cols.index('tweet.created_at')
cols.insert(created_at_index, cols.pop(cols.index('date')))
X_articles = X_articles[cols]

# Convert 'tweet.created_at' to datetime and sort by 'tweet.created_at'
X_articles['tweet.created_at'] = pd.to_datetime(X_articles['tweet.created_at'])
X_articles = X_articles.sort_values(by='tweet.created_at').reset_index(drop=True)
# Add a constant column named 'source' with value 'X'




# Display the first row to verify
X_articles.head(1)
```

# Renaming Columns so that they match 

```python
# Rename 'pub_date' to 'date_and_time'
NYT_articles = NYT_articles.rename(columns={'pub_date': 'date_and_time'})
NYT_articles = NYT_articles.rename(columns={'headline': 'text'})
NYT_articles = NYT_articles.rename(columns={'byline': 'by'})
```

```python
X_articles = X_articles.rename(columns={'tweet.created_at': 'date_and_time'})
X_articles = X_articles.rename(columns={'tweet.text': 'text'})
X_articles = X_articles.rename(columns={'tweet.user.name': 'by'})
```

## Filtering tweets with only > 2000 views

```python
# Filter the dataframe to only include rows with 2000 or more views
# Filter the X_articles dataset to keep only rows where 'tweet.view_count' is >= 2000 and reset the index
X_articles_over_2000views = X_articles[X_articles['tweet.view_count'] >= 2000].reset_index(drop=True)


X_articles_over_2000views.tail(1)
```

# Dropping not overlapping columns so that we can merge the X and NYT

```python
# Drop unnecessary columns

NYT_articles = NYT_articles.drop(columns=['article_count','multimedia_url', 'web_url', 'section_name', 'lead_paragraph', 'abstract'])
NYT_articles.head(1)
```

```python
# Drop unnecessary columns

X_articles_over_2000views = X_articles_over_2000views.drop(columns=['tweet_count','tweet.media','tweet.user.followers_count','tweet.favorite_count','tweet.retweet_count','tweet.reply_count','tweet.user.location','tweet.view_count'])
X_articles_over_2000views.head(1)   
```

```python
X_articles_over_2000views.tail(1)
```

```python
NYT_articles.tail(1)
```

# Merge the two datasets

```python

data = pd.merge(NYT_articles, X_articles_over_2000views,how='outer')
# Reordering the columns so that 'publish date' is the first column
cols = ['date'] + [col for col in data.columns if col != 'date']
data = data[cols]

data = data.sort_values(by='date_and_time', ascending=True,).reset_index(drop=True)
data
```

```python
# Save the merged DataFrame to a new CSV file
output_path = 'Merged NYT & X (>2000) views/Merged NYT & X (>2000) views.csv'
data.to_csv(output_path, index=False)

print(f"The merged DataFrame has been saved to {output_path}")
```
