```python
import pandas as pd

file_path = 'NYT_articles_results.csv'

# Read the CSV files
NYT_articles_results = pd.read_csv(file_path)

NYT_articles_results.tail()
```

```python
import pandas as pd

nyt_articles=NYT_articles_results

# Convert 'pub_date' to datetime and create 'pub_date_by_day' column
nyt_articles['pub_date'] = pd.to_datetime(nyt_articles['pub_date'])
nyt_articles['pub_date_by_day'] = nyt_articles['pub_date'].dt.date
cols = list(nyt_articles.columns)
pub_date_index = cols.index('pub_date')
cols.insert(pub_date_index, cols.pop(cols.index('pub_date_by_day')))
nyt_articles = nyt_articles[cols]

nyt_articles
```

# Calculate the mean of compound sentiment scores grouped by 'pub_date_by_day'
We have multiple tweets in a day, therefore we need mean for the sentiment score each day.

```python
# Calculate the mean of compound sentiment scores grouped by 'pub_date_by_day'
compound_mean_by_day = nyt_articles.groupby('pub_date_by_day')['compound(sentiment_score_headline)'].mean().reset_index()
compound_mean_by_day.columns = ['pub_date_by_day', 'compound_mean_by_day']  

# Merge the calculated means back into the original dataframe
nyt_articles = pd.merge(nyt_articles, compound_mean_by_day, on='pub_date_by_day', how='left')

nyt_articles
```


## The code below will calculate the overlapping date range between the two datasets (nyt_articles_new and x_articles_df). The start_date will be the later of the two earliest dates, and the end_date will be the earlier of the two latest dates. This ensures that any further analysis can be performed on the date range that both datasets share in common. 

```python
import pandas as pd

# Convert date columns to datetime if they are not already
nyt_articles_new['pub_date_by_day'] = pd.to_datetime(nyt_articles_new['pub_date_by_day'])
x_articles_df['tweet.created_at_by_day'] = pd.to_datetime(x_articles_df['tweet.created_at_by_day'])

# Define the common date range
start_date = max(nyt_articles_new['pub_date_by_day'].min(), x_articles_df['tweet.created_at_by_day'].min())
end_date = min(nyt_articles_new['pub_date_by_day'].max(), x_articles_df['tweet.created_at_by_day'].max())

# Filter the dataframes to this common date range
nyt_filtered = nyt_articles_new[(nyt_articles_new['pub_date_by_day'] >= start_date) & (nyt_articles_new['pub_date_by_day'] <= end_date)]
tweets_filtered = x_articles_df[(x_articles_df['tweet.created_at_by_day'] >= start_date) & (x_articles_df['tweet.created_at_by_day'] <= end_date)]

# Plotting the aligned data
sns.set_style("darkgrid", {"axes.facecolor": ".9"})

plt.figure(figsize=(14, 7))

# Plot: NYT data
sns.lineplot(x='pub_date_by_day', y='compound_mean_by_day', data=nyt_filtered, marker='o', linewidth=2.5, color='blue', label='NYT Data')

# Plot: Tweet data
sns.lineplot(x='tweet.created_at_by_day', y='compound_mean_by_day', data=tweets_filtered, marker='o', linewidth=2.5, color='orange', label='Tweet Data')

# Adding titles and labels
plt.title('Compound Sentiment Score by Day (NYT & Tweet Data)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Compound Sentiment Score', fontsize=14)
plt.xticks(rotation=45)
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

## We create all sorts of "difference" columns, as it is are goal to find cause and effect. 
### We normalize them as well

```python
# Calculate the difference between 'Open' and 'Close' prices
oil_data['Open_Close_Diff'] = oil_data['Open'] - oil_data['Close']

# Normalize the 'Open_Close_Diff' column
min_diff = oil_data['Open_Close_Diff'].min()
max_diff = oil_data['Open_Close_Diff'].max()
oil_data['Normalized_Open_Close_Diff'] = (oil_data['Open_Close_Diff'] - min_diff) / (max_diff - min_diff)


oil_data['High_Open_Diff'] = oil_data['High'] - oil_data['Open']


min_diff = oil_data['High_Open_Diff'].min()
max_diff = oil_data['High_Open_Diff'].max()
oil_data['Normalized_High_Open_Diff'] = (oil_data['High_Open_Diff'] - min_diff) / (max_diff - min_diff)


oil_data['Low_Open_Diff'] = oil_data['High'] - oil_data['Open']

min_diff = oil_data['Low_Open_Diff'].min()
max_diff = oil_data['Low_Open_Diff'].max()
oil_data['Normalized_Low_Open_Diff'] = (oil_data['Low_Open_Diff'] - min_diff) / (max_diff - min_diff)


min_diff = oil_data['Volume'].min()
max_diff = oil_data['Volume'].max()
oil_data['Normalized_Volume'] = (oil_data['Volume'] - min_diff) / (max_diff - min_diff)


oil_data.head()


```
